```
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm
```

#List 1


## Exercise 1: (Rule of 12)

###Generate vector of random variables

```
def generate_z_vect(n_u, vect_len):
  EU = 0.5
  VarU = 1/12
  U = np.random.uniform(size=(n_u, vect_len))
  Z = (np.sum(U, axis=0) - 12*EU) / (np.sqrt(12*VarU))
  return Z
```

```
vect_z = generate_z_vect(12, 10000)
```

###Compare CDFs of $Z$ and $N(0,1)$

```
rv = norm()
x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), 100)
```

```
fig, ax = plt.subplots(figsize=(15,10))
sns.lineplot(x=x, y=norm.cdf(x), ax=ax)
sns.ecdfplot(vect_z, ax=ax)
plt.title('Comparison of $CDFs$')
plt.show()
```

###Compare PDFs of $Z$ and $N(0,1)$

```
fig, ax = plt.subplots(figsize=(15,10))
sns.lineplot(x=x, y=norm.pdf(x), ax=ax)
sns.kdeplot(vect_z, ax=ax)
plt.title('Comparison of $PDFs$')
plt.show()
```
