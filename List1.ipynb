```
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm
```

## Exercise 1: (Rule of 12)

###Generate vector of random variables

```
def generate_z_vect(n_u, vect_len):
  EU = 0.5
  VarU = 1/12
  U = np.random.uniform(size=(n_u, vect_len))
  Z = (np.sum(U, axis=0) - 12*EU) / (np.sqrt(12*VarU))
  return Z
```

```
vect_z = generate_z_vect(12, 10000)
```

```
def generate_comparison_plot(z_vector, type_, x_min, x_max, y_scale='linear', x_scale='linear'):
  rv = norm()
  x = np.linspace(x_min, x_max, 1000)
  theory_func = norm.cdf if type_ == 'CDF' else norm.pdf if type_ == 'PDF' else 0
  plot_empirical_func = sns.ecdfplot if type_ == 'CDF' else sns.kdeplot if type_ == 'PDF' else 0

  fig, ax = plt.subplots(figsize=(12,8))
  sns.lineplot(x=x, y=theory_func(x), ax=ax)
  plot_empirical_func(z_vector, ax=ax)
  plt.title(f'Comparison of ${type_}s$')
  plt.yscale(y_scale)
  plt.xscale(x_scale)
  plt.show()
  return fig
```

###Compare CDFs of $Z$ and $N(0,1)$

```
cdf_comparison = generate_comparison_plot(vect_z, 'CDF', -5, 5)
```

```
cdf_comparison_log_scales = generate_comparison_plot(vect_z, 'CDF', -4, 4, 'log')
```

###Compare PDFs of $Z$ and $N(0,1)$

```
pdf_comaprison = generate_comparison_plot(vect_z, 'PDF', -5, 5)
```

```
cdf_comparison_log_scales = generate_comparison_plot(vect_z, 'PDF', -4, 4, 'log')
```
